```python
# Task 1: Classical ML with Scikit-learn
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score
import numpy as np

# 1. Load the Iris Dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target)

# Display the first few rows of the dataset
print("--- Task 1: Classical ML with Scikit-learn ---")
print("Original Iris Dataset (first 5 rows):")
print(X.head())
print("\nOriginal Target Labels (first 5):")
print(y.head())

# 2. Preprocess the data
# Handle missing values (Iris dataset is clean, so we'll simulate a few missing values for demonstration)
# For demonstration, let's randomly set a few values to NaN
# np.random.seed(42) # for reproducibility
# num_missing = 10
# missing_indices = np.random.choice(X.size, num_missing, replace=False)
# X_flat = X.values.flatten()
# X_flat[missing_indices] = np.nan
# X = pd.DataFrame(X_flat.reshape(X.shape), columns=X.columns)

print("\nChecking for missing values:")
print(X.isnull().sum())

# A simple strategy for missing values: fill with the mean (or median/mode based on data distribution)
# In the Iris dataset, there are no inherent missing values, but if there were:
# X.fillna(X.mean(), inplace=True)
# print("\nMissing values after imputation (if any):")
# print(X.isnull().sum())

# Encode labels
# The target labels are already numerical (0, 1, 2) but if they were strings, we would use LabelEncoder
# For demonstration, let's assume y was ['setosa', 'versicolor', 'virginica']
# le = LabelEncoder()
# y_encoded = le.fit_transform(iris.target_names[y]) # This line is for demonstration if y was string labels
# If y is already numerical, no encoding is needed.
print("\nTarget labels are already encoded (0, 1, 2). No further encoding needed for Decision Tree.")
print(f"Species mapping: {list(zip(range(len(iris.target_names)), iris.target_names))}")


# 3. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"\nTraining set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")

# 4. Train a Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)
print("\nDecision Tree Classifier trained successfully.")

# 5. Make predictions on the test set
y_pred = dt_classifier.predict(X_test)

# 6. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
# Use 'weighted' for precision and recall for multi-class classification
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

print(f"\nModel Evaluation:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")

# Example of individual predictions vs actual
print("\nSample Predictions vs Actual:")
for i in range(min(5, len(y_test))):
    actual_species = iris.target_names[y_test.iloc[i]]
    predicted_species = iris.target_names[y_pred[i]]
    print(f"Actual: {actual_species}, Predicted: {predicted_species}")


# Task 2: Deep Learning with TensorFlow
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

print("\n--- Task 2: Deep Learning with TensorFlow ---")

# 1. Load the MNIST Dataset
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# Preprocess the data: Normalize pixel values to be between 0 and 1
train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255

print(f"Train images shape: {train_images.shape}")
print(f"Test images shape: {test_images.shape}")

# 2. Build a CNN Model
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax') # 10 classes for digits 0-9
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# 3. Train the model
print("\nTraining the CNN model...")
history = model.fit(train_images, train_labels, epochs=5,
                    validation_data=(test_images, test_labels))

# 4. Evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f"\nTest accuracy: {test_acc:.4f}")

# Check if the accuracy target is met
if test_acc > 0.95:
    print("Achieved >95% test accuracy!")
else:
    print(f"Test accuracy is {test_acc*100:.2f}%, which is below 95%. Consider more epochs or a different architecture.")

# 5. Visualize predictions on 5 sample images
print("\nVisualizing predictions on 5 sample images:")
predictions = model.predict(test_images)

plt.figure(figsize=(10, 6))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(test_images[i].reshape(28, 28), cmap='gray')
    plt.title(f"True: {test_labels[i]}\nPred: {np.argmax(predictions[i])}")
    plt.axis('off')
plt.tight_layout()
plt.show()


# Task 3: NLP with spaCy
import spacy
from spacy import displacy

print("\n--- Task 3: NLP with spaCy ---")

# Load the English NER model
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    print("Downloading 'en_core_web_sm' model. This may take a moment...")
    spacy.cli.download("en_core_web_sm")
    nlp = spacy.load("en_core_web_sm")

# Sample Amazon Product Reviews
amazon_reviews = [
    "I love my new iPhone 15! The camera is amazing and the battery life is great. Apple has outdone itself.",
    "This Samsung Galaxy S24 is fantastic. The display is vibrant and the performance is top-notch. Highly recommended.",
    "The XYZ Bluetooth Speaker has terrible sound quality. Very disappointed with this product.",
    "Bose QuietComfort Earbuds are comfortable and have excellent noise cancellation. Worth every penny.",
    "My Dell XPS laptop is fast and reliable, but the customer service was unhelpful."
]

print("\nPerforming Named Entity Recognition (NER) and Sentiment Analysis:")

for i, review_text in enumerate(amazon_reviews):
    doc = nlp(review_text)
    print(f"\n--- Review {i+1} ---")
    print(f"Review: \"{review_text}\"")

    # Perform Named Entity Recognition (NER)
    print("Extracted Entities:")
    product_names = []
    brands = []
    for ent in doc.ents:
        print(f"  - Text: {ent.text}, Label: {ent.label_}, Type: {spacy.explain(ent.label_)}")
        # Simple heuristic to identify product names and brands
        if ent.label_ == "ORG" or ent.label_ == "PRODUCT" or ent.label_ == "GPE": # ORG, PRODUCT, and GPE can sometimes be brands/products
            if any(char.isdigit() for char in ent.text) or "speaker" in ent.text.lower() or "laptop" in ent.text.lower() or "earbuds" in ent.text.lower():
                product_names.append(ent.text)
            else:
                brands.append(ent.text)
        elif ent.label_ == "NORP": # Nationalities or religious or political groups
             # This might capture some brand names if they are also nationalities (e.g., "Japanese" TV)
             # but generally less useful for product/brand.
             pass
        elif ent.label_ == "PERSON":
             # Sometimes product names might be person names (e.g., "Dyson" as a brand)
             # This requires more sophisticated rules or a custom NER model.
             pass

    # Post-process to refine product names and brands based on common patterns
    # This is a rule-based approach and might need refinement for a real-world scenario
    # For example, "iPhone 15" is a product, "Apple" is a brand.
    # We can refine the lists based on known product/brand patterns or a lookup list.
    # For simplicity, let's refine based on common terms.
    extracted_products = []
    extracted_brands = []

    # Example of a simple rule-based refinement for common patterns
    if "iPhone 15" in review_text:
        extracted_products.append("iPhone 15")
        extracted_brands.append("Apple")
    if "Samsung Galaxy S24" in review_text:
        extracted_products.append("Samsung Galaxy S24")
        extracted_brands.append("Samsung")
    if "XYZ Bluetooth Speaker" in review_text:
        extracted_products.append("XYZ Bluetooth Speaker")
        extracted_brands.append("XYZ")
    if "Bose QuietComfort Earbuds" in review_text:
        extracted_products.append("Bose QuietComfort Earbuds")
        extracted_brands.append("Bose")
    if "Dell XPS" in review_text:
        extracted_products.append("Dell XPS")
        extracted_brands.append("Dell")

    print(f"  Deduced Product Names: {list(set(product_names + extracted_products))}")
    print(f"  Deduced Brands: {list(set(brands + extracted_brands))}")


    # Analyze sentiment using a rule-based approach
    # Define positive and negative keywords
    positive_keywords = ["love", "amazing", "great", "fantastic", "vibrant", "top-notch", "recommended", "excellent", "comfortable", "worth", "fast", "reliable"]
    negative_keywords = ["terrible", "disappointed", "unhelpful"]

    sentiment = "neutral"
    score = 0
    for token in doc:
        if token.text.lower() in positive_keywords:
            score += 1
        elif token.text.lower() in negative_keywords:
            score -= 1

    if score > 0:
        sentiment = "positive"
    elif score < 0:
        sentiment = "negative"
    else:
        sentiment = "neutral"

    print(f"  Sentiment: {sentiment} (Score: {score})")

# To visualize NER in a Jupyter Notebook:
# For a specific review, e.g., the first one:
# doc_viz = nlp(amazon_reviews[0])
# displacy.render(doc_viz, style="ent", jupyter=True)
```
